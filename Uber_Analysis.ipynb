```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Tasks/ncr_ride_bookings.csv')
```

```python
# Explore dataset: check shape, columns, data types, and first few rows.
print('Shape of df: ',df.shape)
print('coulmns of df:',df.columns)
print('Data types of df: ',df.dtypes)
df.head()
```

```python
#removing the unnecessary things
df['Customer ID'] = df['Customer ID'].str.replace('"','')
df['Booking ID'] = df['Booking ID'].str.replace('"','')

df.head(5)
```

```python
#renaming columns
df.columns = ['Date','Time','Booking_ID'
,'Booking_Status'
,'Customer_ID'
,'Vehicle_Type'
,'Pickup_Location'
,'Drop_Location'
,'Avg_VTAT'
,'Avg_CTAT'
,'Cancelled_Rides_Customer'
,'Customer_Cancellation_Reason'
,'Cancelled_Rides_Driver'
,'Driver_Cancellation_Reason'
,'Incomplete_Rides'
,'Incomplete_Rides_Reason'
,'Booking_Value'
,'Ride_Distance'
,'Driver_Ratings'
,'Customer_Rating'
,'Payment_Method']

df.head()
```

```python
##cheaning the data and null values

# Phase 2: Data Cleaning

# Handle missing values (drop, fill, or replace).

# Convert date/time columns into proper datetime format.

# Extract useful features from datetime (hour, day, weekday, month).

# Standardize categorical values if needed (e.g., location names).

# 1. Handle Duplicates
df.duplicated().sum() ## no duplicates


##4. Handle Missing Values
df.isnull().sum()


nan_cols = ['Customer_Cancellation_Reason','Cancelled_Rides_Driver','Driver_Cancellation_Reason','Incomplete_Rides','Incomplete_Rides_Reason','Payment_Method']

df[nan_cols] = df[nan_cols].fillna('Unknown')


num_cols = ['Booking_Value','Ride_Distance','Driver_Ratings','Customer_Rating','Avg_VTAT','Avg_CTAT','Cancelled_Rides_Customer']

df[num_cols] = df[num_cols].fillna(0)
```

```python
df.isnull().sum()
df.head()
```

```python
df['Vehicle_Type'].unique()
df['Pickup_Location'].unique()
df['Payment_Method'].unique()
```

```python
# Phase 3: Exploratory Data Analysis (EDA)

# Count total trips per day.
df.groupby('Date').agg({'Booking_ID':'count'})

# Find busiest hours of the day.
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
df['Hour'] =  pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour
by_hour = df.groupby('Hour').agg({'Booking_ID':'count'})
by_hour.sort_values('Booking_ID', ascending=False)  ### SO SIX O CLOCK IS THE BUSIEST

# Compare weekday vs weekend demand.

# Trips per month (seasonal trends).
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] =  pd.to_datetime(df['Date']).dt.month
by_month = df.groupby('Month').agg({'Booking_ID':'count'})
by_month.sort_values('Booking_ID',ascending=False) ### seasonal trends

# Find top pickup locations.
top_pickup_locations = df.groupby('Pickup_Location').agg({'Booking_ID':'count'})
top_pickup_locations.sort_values('Booking_ID',ascending=False) ##Khandsa

# Find top drop-off locations.
top_dropoff_locations = df.groupby('Drop_Location').agg({'Booking_ID':'count'})
top_dropoff_locations.sort_values('Booking_ID',ascending=False) ##Ashram


# Calculate average trip distance/duration (if available).

avg_distance = df['Ride_Distance'].mean()
print(avg_distance) ##16.753167933333334


# Phase 4: Advanced Analysis

# Identify peak hours for different days of the week.
df['Week'] =  pd.to_datetime(df['Date']).dt.day_name()
week = df.groupby(['Week','Hour']).agg({'Booking_ID':'count'})
week.sort_values('Booking_ID',ascending=False)

### GO MINI WAS SHOWING THE MOST CANCELLATIONS DUE TO AC
df['Customer_Cancellation_Reason'].unique()
cancelled_rides  = df[df['Customer_Cancellation_Reason'] == 'AC is not working']
cancelled_rides.groupby('Vehicle_Type').agg({'Booking_ID':'count'})

# df['Driver_Cancellation_Reason'].unique()
# Compare demand trends in different boroughs/areas.

# If data includes fares → analyze average fare by distance & time.

# If multiple years of data → analyze year-over-year growth.

# Check correlation between features (distance vs fare, time vs demand).
```
